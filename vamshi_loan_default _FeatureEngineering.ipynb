```python
#import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#impoting raw data
df_raw = pd.read_csv("Training Data.csv")
df_raw.head()
```

```python
len(df_raw["Profession"].unique())
```

```python
df_raw["Profession"].unique()
```

```python
prof_dict = pd.Series(df_raw["Profession"].unique()).to_dict()
```

```python
prof_dict = {key:value for value,key in prof_dict.items()}
```

```python
df_raw["Profession"] = df_raw["Profession"].map(prof_dict)
```

```python
df_raw.head()
```

```python
CITY_dict = pd.Series(df_raw["CITY"].unique()).to_dict()
```

```python
CITY_dict = {key:value for value,key in CITY_dict.items()}
```

```python
df_raw["CITY"] = df_raw["CITY"].map(CITY_dict)
```

```python
df_raw.head()
```

```python
STATE_dict = pd.Series(df_raw["STATE"].unique()).to_dict()
```

```python
STATE_dict = {key:value for value,key in STATE_dict.items()}
```

```python
df_raw["STATE"] = df_raw["STATE"].map(STATE_dict)
```

```python
df_raw.head()
```

```python
df_raw["STATE"].value_counts()
```

```python
# Handle Married/Single attribute
df_raw["Married/Single"] = pd.get_dummies(df_raw["Married/Single"],drop_first=True)
```

```python
df_raw.head()
```

```python
df_raw = pd.concat([df_raw,pd.get_dummies(df_raw["House_Ownership"],drop_first=True)],axis=1)
```

```python
df_raw.head()
```

```python
# Handle Car_Ownership attribute
df_raw["Car_Ownership"] = pd.get_dummies(df_raw["Car_Ownership"],drop_first=True)
```

```python
df_raw.head()
```
